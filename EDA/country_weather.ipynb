```python
import json
import requests
import numpy as np
import pandas as pd
from config import api_key

```

```python
# Read in cleaned coffee production csv
coffee_producing_countries = pd.read_csv('../Data/Cleaned/coffee_production_cleaned.csv')
coffee_producing_countries
```

```python
# Set variables
countries = list(coffee_producing_countries['Country'].unique())
years = list(coffee_producing_countries['Years'].unique())
base_url = 'https://history.openweathermap.org/data/2.5/history/city'

# Dump country data 
country_data = []

# Iterate through country and then years
for country in countries:
    for year in years:
        
        # Set params
        params = {
            "q": country,
            "type": 'hour',
            "start": f'{year}-01-01',
            "end": f'{year}-12-31',
            "appid": api_key
        }
        
        # Make and API request using the params dictionaty
        country_responses = requests.get(base_url, params=params)
        print(country_responses)
        
        
        # Convert the API response to JSON format
        country_responses = country_responses.json()
        print(country_responses)
        break
        
        # Append country data to list 
        country_data.append(country_responses)
        
            
```
